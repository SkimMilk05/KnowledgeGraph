```python
# get page content and titles

page_results = dict()
no_page_results = dict()
icd_redirect = dict()

for code, name in ICD9CODES.items():
    try:
        # get page 
        page = wiki.page(name)
        if "List of ICD-9 codes" in page.title:
            icd_redirect[f'Redirect for {code}: {name}'] = ''
            print(f'ICD redirect for {code}: {name}')
        else:
            page_results[f'{code}: {name}: {page.title}'] = page.content

    except wiki.exceptions.PageError:
        no_page_results[f'{code}: {name}'] = ''
        print(f'No results found for {code}: {name}')
```

```python
# save data
import pickle

with open('wikidicts.pkl', 'wb') as f:
    pickle.dump((page_results, no_page_results, icd_redirect), f)
```
